```python
import boto3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from collections import Counter

# Configurar estilo
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Configurar AWS 
AWS_REGION = 'us-east-1'
TABLE_NAME = 'NotesTable'

print("Librerías importadas correctamente")
```

```python
# Conectar a DynamoDB
dynamodb = boto3.resource('dynamodb', region_name=AWS_REGION)
table = dynamodb.Table(TABLE_NAME)

# Obtener todos los datos
response = table.scan()
items = response['Items']

# Paginación
while 'LastEvaluatedKey' in response:
    response = table.scan(ExclusiveStartKey=response['LastEvaluatedKey'])
    items.extend(response['Items'])

print(f"Se obtuvieron {len(items)} notas")

# Crear DataFrame
df = pd.DataFrame(items)
df.head()
```

```python
# Procesar fechas
df['dateCreated'] = pd.to_datetime(df['dateCreated'])
df['date'] = df['dateCreated'].dt.date
df = df.sort_values('dateCreated')

print(f"Rango: {df['date'].min()} a {df['date'].max()}")
print(f"\nDistribución de sentimientos:")
print(df['sentiment'].value_counts())
```

```python
# Histograma de notas por día
notes_per_day = df.groupby('date').size().reset_index(name='count')

plt.figure(figsize=(14, 6))
plt.bar(notes_per_day['date'].astype(str), notes_per_day['count'], 
        color='steelblue', edgecolor='black', alpha=0.7)
plt.xlabel('Fecha', fontsize=12, fontweight='bold')
plt.ylabel('Cantidad de Notas', fontsize=12, fontweight='bold')
plt.title('Histograma: Cantidad de Notas por Día', fontsize=16, fontweight='bold')  
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Total de notas: {df.shape[0]}")
print(f"Días con actividad: {len(notes_per_day)}")
```

```python
# Gráfico circular de sentimientos
sentiment_counts = df['sentiment'].value_counts()

sentiment_emojis = {
    'happy': ':)',
    'sad': ':(',
    'neutral': ';|',
    'angry': '>:('
}

# Labels sin emojis para matplotlib
labels = [s.capitalize() for s in sentiment_counts.index]

colors = {
    'happy': '#FFD93D',
    'sad': '#6BCB77',
    'neutral': '#95BDFF',
    'angry': '#FF6B6B'
}
color_list = [colors.get(s, '#CCCCCC') for s in sentiment_counts.index]

plt.figure(figsize=(10, 8))
plt.pie(sentiment_counts.values, labels=labels, autopct='%1.1f%%', 
        startangle=90, colors=color_list, textprops={'fontsize': 12, 'fontweight': 'bold'},
        explode=[0.05] * len(sentiment_counts))
plt.title('Proporción de Sentimientos en las Notas', fontsize=16, fontweight='bold', pad=20)  # Sin emoji
plt.axis('equal')
plt.tight_layout()
plt.show()

# Mostrar resumen CON emojis en consola (aquí sí funcionan)
print("\n Resumen de Sentimientos:")
for sentiment, count in sentiment_counts.items():
    percentage = (count / df.shape[0]) * 100
    emoji = sentiment_emojis.get(sentiment, '')
    print(f"{emoji} {sentiment.capitalize()}: {count} notas ({percentage:.1f}%)")
```
